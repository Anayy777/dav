```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml

# 1. Load Titanic data from sklearn library
# We use fetch_openml to access the Titanic dataset via the sklearn API
titanic_data = fetch_openml('titanic', version=1, as_frame=True)
df = titanic_data.frame

# Data Cleaning: Handle potential issues with types/missing values for plotting
df['survived'] = df['survived'].astype(float)
df['age'] = df['age'].astype(float)
df['fare'] = df['fare'].astype(float)

# Set the visual style
sns.set_theme(style="whitegrid")

# --- Task a: Bar chart for survivors/non-survivors by sex ---
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='survived', hue='sex')
plt.title('Frequency of Survivors and Non-Survivors by Sex')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Passenger Count')
plt.legend(title='Sex')
plt.show()

# --- Task b: Scatter plot for any two selected features ---
# Selecting 'Age' and 'Fare' to see if ticket price correlates with age
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='age', y='fare', hue='survived', alpha=0.6)
plt.title('Scatter Plot: Age vs Passenger Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()

# --- Task c: Compare density distribution for Age and Passenger Fare ---
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Age Density
sns.kdeplot(df['age'].dropna(), fill=True, ax=axes[0], color='blue')
axes[0].set_title('Density Distribution of Age')

# Fare Density
sns.kdeplot(df['fare'].dropna(), fill=True, ax=axes[1], color='green')
axes[1].set_title('Density Distribution of Passenger Fare')

plt.tight_layout()
plt.show()

# --- Task d: Pair plot for pairwise bivariate distribution ---
# Selecting a subset of numerical columns for a cleaner pair plot
subset_df = df[['survived', 'pclass', 'age', 'fare', 'sibsp']].dropna()
sns.pairplot(subset_df, hue='survived', palette='viridis', diag_kind='kde')
plt.suptitle('Pairwise Bivariate Distribution of Titanic Features', y=1.02)
plt.show()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10, 6)) 
ax = fig.add_subplot(1, 1, 1) 
ax.plot([1,2,3], [4,5,6],
        color='steelblue',
        linestyle='--',
        linewidth=2,
        marker='o',
        markersize=8,
        alpha=0.9,   # visibility
        label='Monthly Sales')

ax.set_title('Sales Report', fontsize=16, fontweight='bold')
ax.set_xlabel('Month', fontsize=12)
ax.set_ylabel('Revenue ($)', fontsize=12)
ax.legend(loc='upper right')       # auto from label= params
ax.legend(['Line1', 'Line2'])       # manual
ax.set_xlim(0, 10)
ax.set_ylim(0, 100)
ax.grid(True, linestyle='--', alpha=0.5)
ax.tick_params(axis='x', rotation=45)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# 1. Create the plots and provide labels
plt.plot(x, y1, label='Sine Wave', color='blue', linestyle='--')
plt.plot(x, y2, label='Cosine Wave', color='red', linewidth=2)

# Add titles and axis labels
plt.title("Trigonometric Functions")
plt.xlabel("Time")
plt.ylabel("Amplitude")

# 2. Call the legend function
plt.legend()

# Display the plot
plt.show()
```
