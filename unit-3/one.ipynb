```python
import pandas as pd

codes = pd.Series(['Black', 'Pink', 'Brown'], index=[0, 2, 5])
reindex_codes = codes.reindex(range(8), method='ffill')
print(reindex_codes)

# range(8) creates a new index from 0 to 7

# now we only have index 0 , 2 and 5 so the the method ffill i.e forward fill tells pandas to fill the empty value as the value of the last index
```

```python
bills = pd.Series([500 , 1800 , 2200] , index = ['Electricity' , 'Water' , 'Gas'])


bills = bills * 1.10;

print(bills)
print()
print(bills.idxmax()) # gives minimum bill
print(bills.idxmin()) # gives maximum bill

```

```python
import pandas as pd
# loc vs iLoc

# loc does label based selection whereas iloc does integer based selection

# loc uses the actual name of rows and columns

# iloc uses the numeric position


data = {'CITY' : ['Gwalior' , 'Delhi' , 'Mumbai' , 'Indore']
}
df = pd.DataFrame(data , index= ['one' , 'two' , 'three' , 'four'])

print(df)
print()
print(df.loc['three' , 'CITY'])
print(df.iloc[2 , 0])
```

```python
df_employees = pd.DataFrame({
    'EmployeeID': [101,102,103,104],
    'Name': ['Liam','Olivia','Noah','Emma'],
    'Department': ['HR','IT','Sales','HR']
})
df_projects = pd.DataFrame({
    'ProjectID': ['P1','P2','P3'],
    'LeadID': [102,101,105],
    'Budget': [50000,75000,30000]
})
# a) inner join — only matching records
inner = pd.merge(df_employees, df_projects,
                  left_on='EmployeeID', right_on='LeadID',
                  how='inner')
print(inner)
# b) left join — all employees preserved
left = pd.merge(df_employees, df_projects,
                 left_on='EmployeeID', right_on='LeadID',
                 how='left')
print(left)
```
