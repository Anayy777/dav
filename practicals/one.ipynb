```python
# 2

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from scipy import stats

iris = load_iris()

df = pd.DataFrame(data= iris.data , columns= iris.feature_name)

df['species'] = iris.target

df['species_name']  = df['species'].map({i : name for i , name in enumerate(iris.target_name)})

print("Dataset Preview : ")
print(df.head())

# Select only numerical features
features = iris.feature_names
stats_results = []

for col in features:
    mean = df[col].mean()
    median = df[col].median()
    mode = df[col].mode()[0]
    std_dev = df[col].std()
    
    # Standard Error of the Mean (SEM)
    sem = stats.sem(df[col])
    
    # 95% Confidence Interval (CI)
    # Formula: Mean +/- (t_critical * SEM)
    ci = stats.t.interval(0.95, len(df[col])-1, loc=mean, scale=sem)
    
    stats_results.append({
        'Feature': col,
        'Mean': mean,
        'Median': median,
        'Mode': mode,
        'Std Dev': std_dev,
        'Std Error': sem,
        '95% CI': ci
    })

stats_df = pd.DataFrame(stats_results)
print("\n--- Feature Statistics ---")
print(stats_df)

# Compute Correlation Matrix
corr_matrix = df[features].corr()

# Plot Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Iris Features")
plt.show()

print("\nCorrelation Matrix:")
print(corr_matrix)

# Specifically finding covariance between 'sepal length (cm)' and 'petal length (cm)'
cov_val = df['sepal length (cm)'].cov(df['petal length (cm)'])

print(f"\n--- Covariance ---")
print(f"Covariance between Sepal Length and Petal Length: {cov_val:.4f}")

# Binning Petal Length into 'Short', 'Medium', and 'Long' to create a contingency table
df['petal_length_cat'] = pd.cut(df['petal length (cm)'], bins=3, labels=['Short', 'Medium', 'Long'])

# Build Contingency Table
contingency_table = pd.crosstab(df['species_name'], df['petal_length_cat'])

print("\n--- Contingency Table (Species vs Petal Length Category) ---")
print(contingency_table)
```

```python
from IPython.core.pylabtools import figsize

# 3

import seaborn as sns
import matplotlib as plt

df = sns.load_dataset('titanic')

# bar chart

plt.figure(figsize = (10 , 6))
sns.countplot(data = df , x = 'sex' , hue = 'survived')
plt.title('Frequency of Survivors and Non-Survivors by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title = 'survived' , labels = ['yes' , 'no'])
plt.show()


# scatter

plt.figure(figsize = (8 , 5))\
sns.scatterplot(data = df , x = 'age' , y = 'fare' , hue='pclass')

# c. Compare density distribution for age and passenger fare
plt.figure(figsize=(10, 5))
# We use kdeplot for density distribution
sns.kdeplot(df['age'], label='Age', fill=True)
sns.kdeplot(df['fare'], label='Fare', fill=True)
plt.title('Density Distribution: Age vs Fare')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()

# d. Pair plot for pairwise bivariate distribution
# We select a few numeric columns to keep the plot readable
sns.pairplot(df[['age', 'fare', 'pclass', 'survived']], hue='survived', palette='husl')
plt.suptitle('Pairwise Bivariate Distribution', y=1.02)
plt.show()  

```
