```python
import numpy as np
import pandas as pd
data = {
    'Sales_across_months' : [200 , 180 , 250 ,230 , 270] ,
    'Adv_across_Months' : [50 , 40 , 60 , 55 , 65]  
    }

df = pd.DataFrame(data);

print("Correlation : \n" , df.corr())
print("Covariance : \n" , df.cov())
```
